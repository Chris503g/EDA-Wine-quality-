### Data Cleaning

#### Inputing missing data with mean

```python
#missing data
wine_data.isnull().sum()
```

```python
# impute missing values via the column mean 

from sklearn.impute import SimpleImputer 

# Using DataFrame to drop string columns 'white and red'
wine_data.drop(wine_data.columns[[0, 0]], axis=1, inplace=True)

imr = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imr = imr.fit(wine_data.values)
imputed_data = imr.transform(wine_data.values)
imputed_data
```

```python
wine_data.fillna(wine_data.mean())
```

### Data Visualization 

```python
# Visualising barplot between white and red wine 
print(df_wine['type'].value_counts())
plt.figure(figsize=(7,7))
sns.barplot(x = df_wine['type'].value_counts().index, 
            y = df_wine['type'].value_counts().values,palette = "YlOrRd")

plt.xlabel('type')
plt.ylabel('Frequency')
plt.title('type Bar Plot')
plt.show()
```

```python
# plotting using lmplot 
sns.lmplot(x ='density', y ='pH', data = df_wine)
# Show the plot
plt.show()
```

```python
# draw scatter plot
plt.scatter(x="density", y="pH", data=df_wine, c ="blue") 
# To show the plot
plt.show()
```

```python
# Using pairplot to show relation between all df columns with pH column
for i in range(0, len(wine_data.columns), 5):
    sns.pairplot(data=wine_data,
                x_vars=wine_data.columns[i:i+5],
                y_vars=['pH'])
```
